# Relegation in the EPL

The sorrow of being relegated is maybe more than the joy of winning a title.

On the face of it, it looks like relegation is normal. A team can bounce back. It's just one season away from the big league, right? In reality though, its much different. Some of the things that happens after relegation:

**Best players leave the club** - Think about it. The best (not faithful) players won't want to play in lower leagues. Some mid-table team will acquire services of your best player.

**Becomes difficult to sign good players** - Inverse of the point above, no good player will want to join a team with average players in the second/third division.

**Lack of sqaud depth** - Pretty obvious. If players leave and no one comes in, your squad size is bound to reduce. This takes a toll in the performances. How? In second tier of English football, the English League Championship (formally called First Division), you have two matches every week. With a trimmed squad, fatigue, tiredness and injuries begin to catch up on you.

**Financial problems** - If you finish in relegation places, you lose on quite a lot of money. Ticket prices also need to be dropped in order to attract fans back to the stadium.

**Psychological blow** - Not every team bounces back immediately. Look at Wigan Athletic and Sunderland. Relegation can act as a chain to further relegations.

Relegation is like death. You can't help but weep in sorrow for the club that got relegated.

In this notebook, we will explore what leads teams to be relegated from the most prestigious league in the world using the R package dplyr and I'll share my takeaways at the end on how your club can avoid this dreadful fate. 

#### Load packages

```r
library(readr)
library(dplyr)
```

#### Read csv file

```r
epl <- read_csv("../input/english-premier-league-standings/EPL_standings_2000-2022.csv")
```

#### Create relegated dataframe

```r
relegated <- epl %>%
    filter(Pos >= 18)
```

## Diving into relegated teams data

#### Teams relegated from the EPL since 2000-01 season
Each season, the three worst preforming teams in the EPL are sent packing and relegated to second-tier league of english football. How many teams have been dropped from the EPL and the number of times each of them have been relegated since the 2000-01 EPL season? Lets find out.

```r
relegated %>%
    group_by(Team) %>%
    summarise(Times_relegated = n()) %>%
    arrange(desc(Times_relegated))
```

36 clubs have seen themselves land in the bottom three of the EPL and relegated to the second-tier enhlish league. Both Norwich City and West Bromwich Albion have the dismal feat of being sent packing five times! Current powerhouse (and champions!) Manchester City has been relegated once during that period and other great Premier League teams Leicester City, Newcastle United, and West Ham United have been relegated twice.

#### Averages for teams that were relegated

```r
relegated %>%
    summarise(avg_wins = mean(W),
             avg_draws = mean(D),
             avg_losses = mean(L),
             avg_goalsscored = mean(GF),
             avg_goalsallowed = mean(GA),
             avg_goaldif = mean(GD),
             avg_points = mean(Pts)) %>%
    mutate_if(is.numeric, round)
```

Teams are sent packing for their poor preformance in the EPL. We see relegated teams have on average only seven wins, nine draws, and twenty-two losses. They only accumulate thirty points for the season and score less than a goal-per-game.

#### Averages for teams that were relegated by position

```r
relegated %>%
    group_by(Pos) %>%
    summarise(avg_wins = mean(W),
             avg_draws = mean(D),
             avg_losses = mean(L),
             avg_goalsscored = mean(GF),
             avg_goalsallowed = mean(GA),
             avg_goaldif = mean(GD),
             avg_points = mean(Pts)) %>%
    mutate_if(is.numeric, round)
```

#### Relegation by season

```r
relegated %>%
group_by(Season) %>%
    summarise(avg_wins = mean(W),
             avg_draws = mean(D),
             avg_losses = mean(L),
             avg_goalsscored = mean(GF),
             avg_goalsallowed = mean(GA),
             avg_goaldif = mean(GD),
             avg_points = mean(Pts)) %>%
    mutate_if(is.numeric, round)
```

#### Best and worst preformance for relegated teams by points

```r
relegated[which.max(relegated$Pts),]
relegated[which.min(relegated$Pts),]
```

## 17th place
The most celebrated place by a club in the EPL (kinda) as they are saved from being sent away from the top flight of english soccer and live to play another season in the most pretegious league in the world. We'll now dive into the position seventeen dataset.

#### Create seventeenth place dataframe

```r
seventeenth <- epl %>%
    filter(Pos == 17)
```

#### Teams that placed 17th in the EPL since the 2000-01 season

```r
seventeenth %>%
    group_by(Team) %>%
    summarise(Times_seventeenth = n()) %>%
    arrange(desc(Times_seventeenth))
```

Eighteen teams have placed in this covenant position and Sunderland, Wow! This club has placed seventeenth three times and have staved off being relegated that season (though they have been relegated three times as well.) Other notable teams that have placed here are Everton, Leeds United, and Southampton.

#### Averages for seventeenth place

```r
seventeenth %>%
    summarise(avg_wins = mean(W),
             avg_draws = mean(D),
             avg_losses = mean(L),
             avg_goalsscored = mean(GF),
             avg_goalsallowed = mean(GA),
             avg_goaldif = mean(GD),
             avg_points = mean(Pts)) %>%
    mutate_if(is.numeric, round)
```

Clubs here do everythings to avoid relegation and there are some similarities to relegated teams. Less than ten wins, underneath forty goals scored, and have a high negative goal differential. They place only an average three points ahead of the eighteenth place clubs, which is equvillant to one win. So they have a slim margin of escaping relegation. 

#### Best and worst preformance for seventeenth place by points

```r
seventeenth[which.max(seventeenth$Pts),]
seventeenth[which.min(seventeenth$Pts),]
```

## Top 4 versus relegated clubs
The best clubs in the EPL for that season will be hailed by their fans for achiving victory. The top four in the league standings recieve qualification to the UEFA's Champions League, the most illustrios club tournament in the world. Let's now compare the datasets of the top 4 to relegated teams.

#### Create top four dataframe

```r
top4 <- epl %>%
    filter(Pos <= 4)
```

#### Top four compared to relegated teams

```r
top4 %>%
    summarise(avg_wins = mean(W),
             avg_draws = mean(D),
             avg_losses = mean(L),
             avg_goalsscored = mean(GF),
             avg_goalsallowed = mean(GA),
             avg_goaldif = mean(GD),
             avg_points = mean(Pts)) %>%
    mutate_if(is.numeric, round)

relegated %>%
    summarise(avg_wins = mean(W),
             avg_draws = mean(D),
             avg_losses = mean(L),
             avg_goalsscored = mean(GF),
             avg_goalsallowed = mean(GA),
             avg_goaldif = mean(GD),
             avg_points = mean(Pts)) %>%
    mutate_if(is.numeric, round)
```

Teams in the top four have on average twenty-four wins during the season and ammasing and average of seventy-nine points. Difference in goals is massive between the two, with top four teams scoring on average seventy-five goals while relegated teams only average thirty-five goals. That less than one per game!

## Keys to avoid relegation
After diving into the datasets, here's are some of my key takeaways for clubs to know on how to avoid being dropped from the English Premier League:

1. Ten wins should be enough to avoid relegation
2. Less than thirty points earned throughout the season will all but assure relegation for your club
3. Allowing sixty plus goals hurts your chances of staying in the EPL
4. Scoring over forty goals (averaging over a goal per game) will help your club survive to play another season
